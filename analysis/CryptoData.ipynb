```python
from pyvis.network import Network
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
df = pd.read_csv('../datasets/data.csv', header=None, names=['from', 'to', 'value'])
df.head()
```

```python
net = Network(notebook=True)
net.from_nx(nx.from_pandas_edgelist(df, source='to', target='from', edge_attr='value'))
```

```python
G = nx.from_pandas_edgelist(df, source='to', target='from', edge_attr='value')
plt.figure(figsize=(10,10))
nx.draw(G, with_labels=False, node_size=5)
plt.show()
```

```python
net.show_buttons(filter_=['physics'])
net.show("example.html")
```
